# OR the probability of events:

**Example :In the experiment of rolling the dice once, let the event A be defined as an even number, and event B as the number being greater than 3. Find the probability that the resulting number is greater than 3 or even.?**

# Solution :
#Let two events A and B be defined in the sample space S. The combination of these events, shown as AUB, is the set of outcomes that satisfy either event A or event B or both events together.
# Let's show the basic results that satisfy the events using sets.

# S = {1, 2, 3, 4, 5, 6}
# A= {2, 4, 6}
# B = {4, 5, 6}
# A∩B = {4, 6}
---------------
  # **P(A)= 3/6**
  # **P(B)= 3/6**
  # **P(A ∩ B) = 2/6**
---------------
## *P(A U B)= P(A) + P(B) - P(A ∩ B)*
## *P(A U B)= 3/6 + 3/6 - 2/6 =* **4/6**


---
---

````python
import sys
from math import factorial
from fractions import Fraction
import numpy as np
import random


ZAR = {1, 2, 3, 4, 5, 6}
A= {2, 4, 6}
B = {4, 5, 6} 

PA= Fraction(len(A),len(ZAR))    #3/6
PB= Fraction(len(B),len(ZAR))    #3/6
A_intersection_B= Fraction(len(A.intersection(B)),len(ZAR))    #2/6

A_U_B= (PA + PB - A_intersection_B)  # A U B = 4/6

print("\n A U B=", A_U_B )

print("\n A ∩ B = ", A.intersection(B))

print( "\n A U B = ", A.union(B))
````

# AND the probability of events:
#Example:The sample space formed by the results of the experiments related to a problem

# Are events A and B independent?

##**Solution:**
#Two events are said to be independent if, when an experiment is performed, the probability of occurrence of an event B does not affect the probability of occurrence of event A.

**n(S)=8**

**n(A)=2**

**n(B)=4**

**n(A∩B)=1**
_____________________________
#P(A)= n(A)/n(S) = **2/8**

#P(B)= n(B)/n(S) = **4/8**

#P(A∩B)= n(A∩B)/ n(S) =**1/8**
______________________________
#If events A and B are independent: **P(A∩B)=P(A)*P(B)**

#1/8 = 2/8 * 4/8
#1/8 = 8/64
#1/8 = 1/8
#***is independent.***

````python
import sys
from math import factorial
from fractions import Fraction
import numpy as np
import random


S = {1, 2, 3, 4, 5, 6, 7, 8}
A= {1,2}
B = {2, 4, 6, 8} 

nS=len(S)      #n(S)=8

nA=len(A)      #n(A)

nB=len(B)       #n(B)

nA_intersection_B =1     #n(A∩B)=1

PA= Fraction(nA,nS)     #2/8
PB= Fraction(nB,nS)     #4/8
P_A_intersection_B = Fraction(nA_intersection_B,nS)    #1/8

print("\n PA=", PA , "\n PB=", PB , "\n P(A∩B)=",P_A_intersection_B )


if P_A_intersection_B == PA*PB:
  print("\n P(A∩B)=", P_A_intersection_B , "\n PA*PB =", PA*PB)
  print("\n Bağımsızdır (Is independent)")
 
else:
  print("Bağımsız yoktur (Is not independent)")
````
